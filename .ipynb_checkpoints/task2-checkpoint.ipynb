## Задание 2

```python
t = np.linspace(0,10,51)
f = np.cos(t)
```

```python
# цвет линии - pantone зелёный :)
plt.plot(t,f,color = '#699E6D')

title_font = {
    "fontsize" : 16,
    "fontweight" : "bold",
    "color" : "#000000",
    "family" : "Comic Sans MS"
}

label_font = {
    "fontsize" : 12,
    "fontweight" : "bold",
    "color" : "#000000",
    "family" : "Comic Sans MS"
}

plt.title('График f(t)', fontdict = title_font)
plt.xlabel("Значения t", fontdict=label_font)
plt.ylabel("Значения f", fontdict=label_font)

plt.axis([0.5, 9.5, -2.5, 2.5])

#  fontdict=label_font
# ?plt.plot
```

## Задание 3

```python
x = np.linspace(-3, 3, 51)

y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)

```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)

fig.set_size_inches(8,6)

fig.subplots_adjust(wspace=0.3)
fig.subplots_adjust(hspace=0.3)

ax_1, ax_2, ax_3, ax_4 = ax.flatten()

ax_1.plot(x,y1)
ax_1.set_title('График 1', fontdict = label_font)

ax_2.plot(x,y2)
ax_2.set_title('График 2', fontdict = label_font)

ax_3.plot(x,y3)
ax_3.set_title('График 3', fontdict = label_font)

ax_4.plot(x,y4)
ax_4.set_title('График 4', fontdict = label_font)


```

```python
import pandas as pd
```

```python
plt.style.use('fivethirtyeight')

data = pd.read_csv("..\dataset\creditcard.csv")
```

```python
data.value_counts('Class').plot(kind = 'bar')
```

```python
data.value_counts('Class').plot(kind = 'bar', logy=True)
```

```python
# pd.pivot_table(data, index=['V1'], values = ['Class'])
plt.hist(data.loc[data['Class'] == 1 , 'V1'], density = True,  bins = 20, alpha = 0.5, color = 'red', label = 'Class 1')
plt.hist(data.loc[data['Class'] == 0 , 'V1'], density = True,  bins = 20, alpha = 0.5, color = 'grey', label = 'Class 0')
plt.legend(loc = 'upper left')
plt.ylabel("V1", fontdict=label_font)
plt.show()

                                                  
```
